# PROBLEM STATEMENT
The objective is to identify patterns and insights related to customer demographics, shopping preferences, payment methods, shipping types, and overall purchase trends. This analysis will help optimize marketing strategies, improve customer satisfaction, and drive sales growth.



# Column Description
Customer ID:- Unique identifier for each customer (integer).

Age:- Age of the customer (integer).

Gender:- Gender of the customer (e.g., Male, Female).

Item Purchased:- Specific item purchased by the customer.

Category:- Broad category to which the purchased item belongs (e.g., Clothing, Footwear).

Purchase Amount (USD):- Amount spent by the customer in USD (integer).

Location:- State or region of the customer.

Size:- Size of the item purchased (e.g., S, M, L).

Color:- Color of the purchased item.

Season:- Season during which the item was purchased (e.g., Winter, Spring).

Review Rating:- Customer's rating for the purchased item (float).

Subscription Status:- Whether the customer is subscribed to newsletters or services (e.g., Yes, No).

Payment Method:- Method used for payment (e.g., Credit Card, PayPal, Cash).

Shipping Type:- Shipping option chosen (e.g., Express, Free Shipping).

Discount Applied:- Indicates if a discount was applied (e.g., Yes, No).

Promo Code Used:- Indicates if a promo code was used (e.g., Yes, No).

Previous Purchases:- Number of previous purchases made by the customer.

Preferred Payment Method:- The customer's preferred payment method.

Frequency of Purchases:- How often the customer makes purchases (e.g., Weekly, Fortnightly, Annually).

```python
plt.figure(figsize=(10, 6))
sns.barplot(x="Purchase Amount (USD)", y="Category", data=df, estimator=sum, palette="viridis")
plt.title("Total Purchase Amount by Category", fontsize=16, fontweight="bold")
plt.xlabel("Total Purchase Amount (USD)", fontsize=12)
plt.ylabel("Category", fontsize=12)
plt.grid(axis="x", linestyle="--", alpha=0.7) 
plt.tight_layout()  
plt.show()
```

#### OBSERVATION: 

we can observe that the highest purchase amount is in the Clothing category, followed by Accessories. These categories appear to have significantly higher total purchase amounts compared to others.

```python
plt.figure(figsize=(15,5))
sns.kdeplot(df['Purchase Amount (USD)'], color='skyblue')
plt.title('purchase amount', fontsize=16)
plt.xlabel('Purchase Amount (USD)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

### OBSERVATION:
The distribution appears to be slightly skewed to the right, suggesting that there are some higher-value purchases, but they are less frequent.

```python

# Clustered Bar Chart: Purchase Amount by Category, grouped by Gender
sns.barplot(data=df, x="Category", y="Purchase Amount (USD)", hue="Gender")
plt.title("Clustered Bar Chart: Purchase Amount by Category and Gender")
plt.show()
```

```python
# Prepare data for a stacked bar chart
stacked_data = df.groupby(["Category", "Gender"])["Purchase Amount (USD)"].sum().unstack()

# Stacked Bar Chart
stacked_data.plot(kind="bar", stacked=True, figsize=(10, 6), colormap="viridis")
plt.title("Stacked Bar Chart: Purchase Amount by Category and Gender")
plt.ylabel("Total Purchase Amount (USD)")
plt.show()
```

### OBSERVATION:
The clustered bar chart shows the purchase amount by category, indicating that males tend to spend more in categories like Electronics and Clothing, while females spend more in categories like Beauty and Accessories 

# Key Differences:

### Clustered Bar Charts:

Definition: Groups bars side by side for each category, with a separate bar for each subgroup.

Purpose: Highlights comparisons between subgroups within each main category.

### Stacked Bar Charts:

Definition: Stacks the bars for subgroups on top of each other within each category.

Purpose: Highlights the total value of a category, with contributions of subgroups shown proportionally.

```python
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x="Category", hue="Season", palette="Set2")
plt.title("Count of Purchases by Category and Season")
plt.xlabel("Category")
plt.ylabel("Count of Purchases")
plt.legend(title="Season")
plt.show()
```

### OBSERVATION
It shows that Clothing and Accessories have the highest purchase counts across all seasons. Winter sees a higher number of purchases in categories like Clothing, while Summer has a noticeable increase in Accessories.

```python
data = {
    'Purchase Amount': [100, 150, 200, 250, 300, 350, 400, 450, 500],
    'Review Rating': [4.1, 4.5, 4.7, 4.8, 3.9, 4.3, 4.6, 4.2, 5.0]
}

# Creating DataFrame
df = pd.DataFrame(data)

plt.figure(figsize=(8,6))
plt.scatter(df['Purchase Amount'], df['Review Rating'], color='blue')
plt.title('Purchase Amount vs. Review Rating', fontsize=14)
plt.xlabel('Purchase Amount', fontsize=12)
plt.ylabel('Review Rating', fontsize=12)
plt.grid(True)
plt.show()
```

### OBSERVATION
we can observe a general positive correlation: as the purchase amount increases, the review rating tends to increase as well.

```python
data = {
    'Age': [25, 30, 35, 40, 45, 50, 55, 60],
    'Purchase Amount (USD)': [100, 150, 200, 250, 300, 350, 400, 450],
    'Previous Purchases': [2, 3, 5, 1, 4, 6, 2, 3]
}

df = pd.DataFrame(data)

correlation_matrix = df.corr()
plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Age, Purchase Amount (USD), Previous Purchases', fontsize=14)
plt.show()
```

### OBSERVATION
* Purchase Amount (USD) vs Previous Purchases: A strong positive correlation (0.98), meaning that people who have made more previous purchases tend to spend more.

* Age vs Purchase Amount (USD): A moderate positive correlation (0.96), suggesting older individuals tend to spend more.

* Age vs Previous Purchases: A weak correlation (0.38), indicating that age does not have a strong influence on how many previous purchases someone has made.

```python
df = pd.DataFrame({
    'Purchase Amount': [100, 150, 200, 250, 300, 350, 400, 450],
    'Previous Purchases': [2, 3, 5, 1, 4, 6, 2, 3]
})

plt.pie(df.sum(), labels=df.columns, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'], explode=(0.1, 0))
plt.title('Purchase Amount vs Previous Purchases')
plt.axis('equal') 
plt.show()
```

### OBSERVATION

The chart shows that the Purchase Amount contributes a larger portion of the total, while Previous Purchases account for a smaller percentage. 

```python
data = {
    'Payment_Method': ['Credit Card', 'Debit Card', 'Cash', 'UPI', 'Credit Card', 'Cash'],
    'Purchase_Amount': [120, 250, 80, 300, 150, 60]
}

df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.boxplot(x='Payment_Method', y='Purchase_Amount', data=df, color="yellow")
plt.title('Boxplot of Purchase Amount by Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Purchase Amount')
plt.show()
```

### OBSERVATION

Credit Card and Debit Card have higher median purchase amounts compared to Cash and UPI.


```python
data = {
    'Shipping_Type': ['Standard', 'Express', 'Standard', 'Overnight', 'Express', 'Standard', 'Overnight'],
    'Order_Count': [5, 3, 6, 2, 4, 7, 3]
}

df = pd.DataFrame(data)

shipping_counts = df['Shipping_Type'].value_counts()

plt.figure(figsize=(8, 8))
shipping_counts.plot.pie(autopct='%1.1f%%', colors=sns.color_palette("pastel"), startangle=90)
plt.title('Proportion of Shipping Types')
plt.ylabel('')  
plt.show()
```

### OBSERVATION
Standard Shipping is the most common, accounting for the largest proportion of the orders.

Express Shipping and Overnight Shipping contribute smaller but notable shares, with Express being more frequent than Overnight.
